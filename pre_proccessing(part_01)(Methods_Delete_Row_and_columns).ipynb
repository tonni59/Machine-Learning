```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Dataset**

```python
df = pd.read_csv("/content/train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
null_var = df.isnull().sum()/df.shape[0] * 100
null_var
```

```python
drop_columns = null_var[null_var > 20].keys()
drop_columns
```

```python
df2_drop_clm = df.drop(columns=drop_columns)
df2_drop_clm.shape
```

```python
sns.heatmap(df2_drop_clm.isnull())
```

```python
df3_drop_rows = df2_drop_clm.dropna()
```

```python
df3_drop_rows.shape
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_drop_rows.isnull())
```

```python
df3_drop_rows.isnull().sum()
```

```python
df3_drop_rows.select_dtypes(include=['int64','float64']).columns
```

```python
sns.distplot(df['Age'])
```

```python
sns.distplot(df3_drop_rows['Age'])
```

```python
sns.distplot(df['Age'])
sns.distplot(df3_drop_rows['Age'])
```

```python
num_var = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']


plt.figure(figsize=(25,25))
for i,var in enumerate(num_var):
  plt.subplot(9,4,i+1)
  sns.distplot(df[var] , bins=20)
  sns.distplot(df3_drop_rows[var], bins=20)
```

```python
df3_drop_rows.select_dtypes(include=['object']).columns
```

```python
pd.concat([df['Embarked'].value_counts()/df.shape[0] * 100 ,
           df3_drop_rows['Embarked'].value_counts()/df3_drop_rows.shape[0] * 100] , axis=1,
           keys=['Embarked_original', 'Embarked_clean'])
```

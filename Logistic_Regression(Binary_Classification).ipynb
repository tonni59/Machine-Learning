```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/insurance (1).csv")
```

```python
dataset
```

```python
new_df = dataset[['age', 'smoker']]
```

```python
new_df
```

```python
new_df['smoker'].replace({'no':'0','yes':'1'},inplace=True)
```

```python
new_df
```

```python
plt.scatter(x='age',y='smoker',data=new_df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_df[['age']],new_df['smoker'],test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```
